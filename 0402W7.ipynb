````python
# Code Example – Using NLTK word_tokenize()
import nltk

# Download the necessary Punkt Sentence Tokenizer data
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize

text = "Hello, world! This is a test."
tokens_nltk = word_tokenize(text)  # or word_tokenize(text)
print("Tokens from word_tokenize():", tokens_nltk)
````

````python
# Code Example – Using spaCy Tokenization
import spacy
# Load English model (ensure you have installed spaCy and the model)
nlp = spacy.load("en_core_web_sm")

text = "Hello, world! This is a test."
doc = nlp(text)
tokens_spacy = [token.text for token in doc]
print("Tokens from spaCy:", tokens_spacy)
````

**Tokenization**
The text is first tokenized using NLTK's word_tokenize(), which breaks the text into words and punctuation tokens.

**Word Frequency Analysis**
Using FreqDist, we count how many times each token appears, which gives us the frequency distribution of words in the text.
**Word Cloud Generation**
We then generate a word cloud using the WordCloud library. The word cloud visually represents the token frequencies; tokens that appear more frequently are displayed in larger fonts.


````python
import nltk
# Download the book resource (if not already downloaded)
nltk.download('book')

from nltk.book import *  # text1, text2, ..., are now available
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# text1 is a list of words from the first text (commonly Moby Dick)
# Calculate the frequency distribution of all words in text1
freq_text1 = FreqDist(text1)

# Print the top 30 most common words in text1
print("Top 30 most common words in text1:")
print(freq_text1.most_common(30))

# Generate a word cloud from the frequency distribution
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_text1)

# Display the word cloud
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud for text1")
plt.show()
````
