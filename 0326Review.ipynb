```python
import string

# 定義標點符號集合
punctuations = set(string.punctuation)

# 範例字串
text = "Hello, world! This is a test: remove punctuation."

# 移除標點符號：逐一檢查字串中的每個字元，若不在標點符號集合中則保留
clean_text = [ch for ch in text if ch not in punctuations]
clean_text =''.join(clean_text)
print(clean_text)
```

```python
from nltk.corpus import stopwords

# 定義停用詞集合（全部是小寫）
stop_words = set(stopwords.words('english'))

# 範例字串
text = "Hello, world! This is a test: remove punctuation."

# 過濾停用詞，將字串轉換成小寫並逐一檢查每個詞
filtered_text = ''.join([word for word in text.lower().split() if word not in stop_words])
print(filtered_text)
```
